# 二次元配列

````python
data = [["+", "-", "*"], ["あ", "い", "う"], ["□", "⚪︎", "△"]]
data = pd.DataFrame(data)
print(data)
````

````python
# + の出力
print(data[0][0])
print(data.at[0, 0]) # 上の別の書き方
````

````python
# あいうの出力
print(data.loc[1,:])
# □⚪︎△の出力
print(data.iloc[2,:])
````

# 3次元の場合

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# データを生成
x = [1,1,2,2]
y = [5,10,10,25]
z = [1,1,1,1]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 点を描く（scatter）
ax.scatter3D(x, y, z, color='red', s=50)  # s=サイズ

plt.show()
````

````python
# データを生成
x = [1,2]
y = [3,4]
x, y = np.meshgrid(x, y)
print(x)
print(y)
z = np.array([[1, 1], [2, 4]]) # np.ones_like(x)
print(z)
# 3Dプロットの作成
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

plt.show()
````
